```python
# Specify the file path
file_path = r"C:\Users\Bartek\Desktop\Titanic.tsv"

import pandas as pd
import numpy as np
from scipy.stats import zscore
from IPython.display import display
# Load the dataset
data = pd.read_csv(file_path, sep="\t")

# Assess the data quality
print(data.head())    # View the first few rows of data
print(data.describe())   # Get summary statistics for the data
print(data.info())    # Get information about the data types and missing values

# Check for outliers
z_scores = np.abs(zscore(data['PassengerId']))
data = data[(z_scores < 3)]

# Check for duplicates
print(data.duplicated().sum())   # Count the number of duplicate rows
data = data.drop_duplicates()    # Remove the duplicate rows

# Check for missing data and NaNs
print(data.isnull().sum())   # Count the number of missing values
data = data.dropna()    # Remove the rows with missing values

# Clean and repair the dataset
data['PassengerId'] = data['PassengerId'].replace('wrong_value', 'correct_value')   # Replace the wrong value with the correct value

# Save the cleaned dataset
data.to_csv("cleaned_data.tsv", sep="\t", index=False)




# Print the DataFrame
df = pd.DataFrame(data)
print(df)

'''
# Print the first few rows of data
display(data.head())

# Get summary statistics for the data
display(data.describe())

# Get information about the data types and missing values
display(data.info())

# Count the number of duplicate rows
print("Number of duplicate rows:", data.duplicated().sum())

# Count the number of missing values
display(data.isnull().sum())

# Save the cleaned dataset
data.to_csv("cleaned_data.tsv", sep="\t", index=False)
'''
```
